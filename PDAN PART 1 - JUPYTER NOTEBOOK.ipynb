###  **Linear Regression**

*What is Linear Regression?* "Linear Regression, also known as ordinary least squares, is the simplest and most basic method for regression. It finds the parameters *w* and *b* that minimize the mean squared error between predictions and the true regression targets, *y*, on the training set. In other words, it is a supervised machine learning model that finds the best fit linear line between the independent and dependent variable. The mean squared error is the sum of the squared differences between the predictions and the true values, divided by the number of six samples. Linear regression has no parameters, which is a benefit, but it also has no way to control model complexity". (Müller & Guido, 2017)

###  **Analysis**

There are many different forms of linear regression, two most common are simple and multiple linear regression. For this task we will use the second, *multiple linear regression*. "Multiple linear regression is a similar to simple linear regression, with the purpose of evaluating the linear relationships between a response variable and several explanatory variables. In the real world multiple linear regression is used more frequently than simple linear regression. This is because with most datasets, the data collection has more variables that can be included and taken into account when analyzing data. So, multiple linear regression allows us to evaluate the relationship between two variables, while controlling for the effect of other variables. Multiple linear regression models use the standard formula, *Y = β0+β1X1+β2X2+⋯+βpXp+ϵ*, which is similar to the simple linear regression equation, except that there is more than one independent variables."(A.Soetewey, 2021)

### Setting Up the X and the Y

Here we set up our X and Y, the purpose for this is for future use in splitting the data.

```python
# WE LET X EQUAL TO ALL THE X VALUES, THAT'S WHY WE DROP PRICE THE Y VARIABLE
X = data_drop.drop('Price', axis=1) 
X
```

```python
# PRICE IS THE Y VARIABLE
Y = data_drop['Price']
Y
```

### Splitting the Data for Training and Testing

We split the dataset into two pieces, one for training and the other for testing, using the X and Y variables we just declared.

```python
# FIRST WE IMPORT THE train_test_split LIBRARY
from sklearn.model_selection import train_test_split # THIS IMPORT IS USED IN SPLITING THE DATA FOR TESTING AND TRAINING

# THIS SPLITS THE DATA, test_size=0.3 OR 30%, AND THE TRAIN SIZE IS EQUAL TO THE REMAINING 70%
# WHY random_state=42, IS TO ENSURE THAT THE OUTPUT THAT IS RUN EACH TIME EQUALS TO THE FIRST RUN OUTPUT WHEN SPLITTING
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
# DISPLAY X_train
X_train
```

```python
# DISPLAY Y_train
Y_train
```

```python
# DISPLAY X_test
X_test
```

### Bringing in the Model (Multiple Linear Regression)

This lets us use the linear regression model in our dataset.

```python
# FIRST WE IMPORT THE LIBRARY
from sklearn.linear_model import LinearRegression # THIS IMPORT FITS THE LINEAR MODEL WITH THE COEFFICIENTS

# DECLARES A VARIABLE THAT'LL USE LINEAR REGRESSION
model = LinearRegression()
```

```python
# TRAINING THE MODEL USING THE X_train AND Y_train
model.fit(X_train, Y_train)
```

### Making Predictions

Here we make predictions using the X_test, which will be used in the analysis and displaying linear regression in graph form

```python
# WE DECLARE A VARIABLE FOR MAKING PREDICTIONS WITH THE X_test
predictions = model.predict(X_test)

# DISPLAY predictions
predictions
```

```python
# NOW WE VISUALISE OUR PREDICTIONS
# DECLARE VARIABLE x_values FOR OUR X VALUES
data_drop['x_values'] = data_drop['Sale'] + data_drop['Weight'] + data_drop['Resolution'] + data_drop['PPI'] + data_drop['Cpu_Core'] + data_drop['Cpu_Freq']
+ data_drop['Internal_Memory'] + data_drop['RAM'] + data_drop['Rear_Camera'] + data_drop['Front_Camera'] + data_drop['Battery'] + data_drop['Thickness'] 

# THIS DISPLAYS OUR PREDICTION IN THE FORM OF A REGPLOT GRAPH
sns.scatterplot(x='x_values', y='Price', data=data_drop)
plt.plot(predictions, color='red')
```

### Evaluating the Model

We do this to see the accuracy of the model we created by calculating the mean, mean absolute error, mean squared error, and root mean squared error.

```python
# IMPORT THE LIBRARIES
# IMPORT mean_absolute_error IMPORT TO CALCULATE MEAN ABSOLUTE ERROR REGRESSION LOSS
# IMPORT mean_squared_error DOES MEAN SQUARED ERROR REGRESSION LOSS
from sklearn.metrics import mean_absolute_error, mean_squared_error 
```

```python
# FIRST WE DETERMINE THE MEAN OF PRICE(Y)
data_drop['Price'].mean()
```

```python
# CALCULATE MEAN ABSOLUTE ERROR
mean_absolute_error(Y_test, predictions)
```

```python
# CALCULATE MEAN SQUARED ERROR
mean_squared_error(Y_test, predictions)
```

```python
# CALCULATE ROOT MEAN SQUARED ERROR
np.sqrt(mean_squared_error(Y_test, predictions))
```
